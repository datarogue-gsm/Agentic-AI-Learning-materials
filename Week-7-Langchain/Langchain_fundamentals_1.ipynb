## 1. Message types

LangChain models operate on a list of messages, not a single string. Understanding these three message types is fundamental to everything else.

````python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
 
# SystemMessage — sets the LLM's behaviour/persona
system = SystemMessage(content='You are a concise technical assistant.')
 
# HumanMessage — the user's input
human = HumanMessage(content='What is LangChain?')
 
# AIMessage — the model's response (used in conversation history)
ai = AIMessage(content='LangChain is a framework for building LLM apps.')
print(system)
print(human)
print(ai)
````

## 3. invoke() vs stream() vs batch()

````python
# invoke() — single synchronous call
result = llm.invoke('Hello')
````

````python
# stream() — token-by-token streaming (great for UI)
for chunk in llm.stream('Explain LangChain in simple terms in 100 characters.'):
    print(chunk.content, end='', flush=True)
````

````python
# batch() — multiple calls in parallel
results = llm.batch(['What is AI?', 'What is ML?', 'What is DL?'])
for r in results:
    print(r.content)
````

# Prompt Formatting / Engineering

💡 **Key Insight:**  
Same model. Same weights. Completely different output — based purely on how you phrase the input.

---

- 🔴 **Prompt A (Vague)**  
  Tell me about neural networks.

- 🟢 **Prompt B (Precise)**  
  You are a machine learning tutor. Explain how a feedforward neural network works to a software engineer with no ML background.  
  Use a real-world analogy and keep it under **150 words**.

## Zero shot prompting 

- Zero-shot prompting gives the model a task with no examples. 
- You rely entirely on the model's pre-trained knowledge and your instruction quality. 
- Works well for common tasks where the model has strong priors.
- Best for:  Classification, translation, summarisation, simple Q&A — tasks the model has seen extensively during training.
- Limitation:  For niche domains or unusual output formats, zero-shot can miss without examples to anchor the pattern.


````python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
 
# Initialise the model
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0.7)
 
response = llm.invoke("""
                      Classify the following customer message as: 
                      billing / technical / general
                      Message: 'I was charged twice for my subscription last month.' 
                      Return ONLY the category label.
""")
print(response.content)
````

## Few shot prompting 
- Few-shot prompting includes 2-5 input-output examples before the actual task. 
- This dramatically anchors the model to the expected output format, tone, and reasoning style.
- Best for:  Custom output formats, domain-specific tasks, consistent tone/style, structured data extraction.
- Pro tip:  Use diverse examples — include edge cases and tricky inputs so the model learns robustness, not just the easy cases.


````python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage
 
# Initialise the model
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0.7)
 
response = llm.invoke("""
                      Extract the action item and owner from meeting notes.
                      Input:  'John will send the revised proposal by Friday.'
                      Output: {"action": "Send revised proposal", "owner": "John", "due": "Friday"}
                      Input:  'Sarah to schedule the client onboarding call.'
                      Output: {"action": "Schedule client onboarding", "owner": "Sarah", "due": null}
                      Input:  'Team to review the Q3 metrics before Monday.'
                      Output: {"action": "Review Q3 metrics", "owner": "Team", "due": "Monday"}
                      Now extract from:
                      Input:  'Mike needs to update the API documentation by next week.'
                      Output:
""")
print(response.content)
````

## Chain-of-Thought (CoT) Prompting

- Chain-of-thought prompting asks the model to reason step by step before giving a final answer. 
- This dramatically improves accuracy on tasks requiring logic, maths, or multi-step reasoning.
- Best for:  Maths, logic puzzles, multi-step reasoning, code debugging, complex classification with justification.
- Why it works:  Forcing intermediate reasoning steps reduces the chance the model jumps to a wrong answer. The scratchpad itself improves accuracy.


````python
# Implicit CoT — just add 'think step by step'
response = llm.invoke("""
                        A train leaves City A at 9:00 AM travelling at 80 km/h.
                        Another train leaves City B (320 km away) at 10:00 AM at 100 km/h.
                        At what time do they meet?
                        
                        Think through this step by step before giving your final answer.

""")
print(response.content)
````

````python
# Explicit CoT — show the reasoning pattern via examples
response = llm.invoke("""
                        Q: If a shirt costs $40 after a 20% discount, what was the original price?
                        A: Let me think step by step.
                        Step 1: $40 is 80% of the original price (100% - 20% = 80%).
                        Step 2: Original price = $40 / 0.80 = $50.
                        Final answer: $50.
                        
                        Q: A laptop costs $900 after a 25% discount. What was the original price?
                        A:
""")
print(response.content)
````

## Schema-Guided prompting

- Schema-guided prompting provides a strict output schema — either as a JSON template, Pydantic model, or typed specification — that the model must conform to. 
- This is the production-grade approach for any system that needs to parse LLM output programmatically.

````python
# Explicit CoT — show the reasoning pattern via examples
response = llm.invoke("""
                        Extract product details from the review below.
                        Return ONLY a JSON object matching this exact schema:
                        
                        {
                        "product_name": string,
                        "rating": integer between 1-5,
                        "pros": list of strings (max 3),
                        "cons": list of strings (max 3),
                        "recommended": boolean
                        }
                        
                        Review:
                        'The noise-cancelling headphones sound incredible and battery lasts 30hrs.
                        However, the ear cups get uncomfortable after 2 hours. Still, worth buying.'
""")
print(response.content)
````

## 1. Prompt Templates with OpenAI + LangChain

````python
# This script shows:
# 1. Creating reusable prompt templates using LangChain
# 2. Injecting dynamic inputs into templates
# 3. Running prompts with latest OpenAI models using the OpenAI Python SDK

# Install (if needed):
# pip install langchain langchain-core langchain-community langchain-openai openai tiktoken


# ------------------------------------------------------------
# Imports
# ------------------------------------------------------------
from langchain_core.prompts import PromptTemplate   # Updated import for new LC version
from openai import OpenAI                           # New OpenAI SDK (1.x style)


# ------------------------------------------------------------
# OpenAI client initialization
# ------------------------------------------------------------
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"),
                base_url=os.getenv("OPENAI_API_BASE"))  # Replace with your real API key


# ------------------------------------------------------------
# Step 1: Create a reusable prompt template
# ------------------------------------------------------------
summary_template = PromptTemplate(
    input_variables=["text"],
    template="""
You are a helpful AI assistant.

Summarize the following text in 3-4 concise bullet points:

Text:
{text}

Summary:
"""
)


# ------------------------------------------------------------
# Step 2: Format the prompt with dynamic input
# ------------------------------------------------------------
user_text = """
Artificial intelligence is transforming industries worldwide.
Companies use AI for automation, predictive analytics, and personalized user experiences.
The demand for AI engineers and data scientists continues to grow.
"""

# Format the template with user-supplied text
prompt = summary_template.format(text=user_text)


# ------------------------------------------------------------
# Step 3: Call OpenAI model with the formatted prompt
# ------------------------------------------------------------
response = client.chat.completions.create(
    model="gpt-4o-mini",     # Latest small model from OpenAI
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print("\nGenerated Summary:\n")
print(response.choices[0].message.content)
````

````python
# ------------------------------------------------------------
# Step 4: Another reusable prompt template (QA template)
# ------------------------------------------------------------
qa_template = PromptTemplate(
    input_variables=["context", "question"],
    template="""
Use the context to answer the question accurately.

Context:
{context}

Question:
{question}

Answer:
"""
)

# Sample dynamic input
context_text = (
    "Python is an interpreted, high-level programming language used in "
    "AI, data science, automation, and web development."
)

question_text = "Why is Python preferred in AI?"

# Format the QA prompt
prompt2 = qa_template.format(
    context=context_text,
    question=question_text
)

# Call the OpenAI model again
response2 = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": prompt2}]
)

print("\nAnswer from QA Template:\n")
print(response2.choices[0].message.content)
````

## 2. ChatPromptTemplate — System + Human Combos

## 3. MessagesPlaceholder — Dynamic History Injection

MessagesPlaceholder lets you inject a variable-length list of messages (e.g. conversation history) into a fixed prompt structure.

````python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage

prompt = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful assistant.'),
    MessagesPlaceholder(variable_name='history'),   # inject history here
    ('human', '{input}')
])

# Assume this is the conversation history we want to inject (could be from a database, previous interactions, etc.) 
history = [
    HumanMessage(content='My name is Goura.'),
    AIMessage(content='Nice to meet you, Goura!'),
]
 
messages = prompt.format_messages(history=history, input='What is my name?')
response = llm.invoke(messages)
print(response.content)  # → 'Your name is Goura.'
````

# LCEL

- LangChain Expression Language (LCEL) is the declarative way to compose chains using the | (pipe) operator — the same concept as Unix pipes. 
- Each component takes input, transforms it, and passes it to the next.
- **Core Idea:  prompt | llm | parser means: format the prompt → send to LLM → parse the output. That's a complete pipeline in one line.**

````python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
 
# Define each component
prompt  = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful assistant.'),
    ('human',  '{input}')
])
llm     = ChatOpenAI(model='gpt-4o-mini', api_key=os.getenv("OPENAI_API_KEY"), 
                     base_url=os.getenv("OPENAI_API_BASE"), temperature=0)
parser  = StrOutputParser()
 
# Compose with | operator
chain = prompt | llm | parser
 
# Invoke
result = chain.invoke({'input': 'What is LangChain?'})
print(result)   # Plain string output
````

# Output Parser

- Output parsers transform that output into the format your application needs.

## Example - Response Parsing / Structured Output Handling


Goal:

- 1. Ask the model to return clean JSON
- 2. Parse safely using Python    
- 3. Handle malformed JSON

- Requirements: pip install openai langchain-core langchain-community langchain-openai tiktoken



#### Without Parser

````python
from openai import OpenAI
from langchain_core.prompts import PromptTemplate   # Updated import for LangChain 1.x
import json


# ------------------------------------------------------------
# OpenAI client (new SDK)
# ------------------------------------------------------------
# client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))


# ------------------------------------------------------------
# Step 1: Create structured output prompt
# ------------------------------------------------------------
json_template = PromptTemplate(
    input_variables=["text"],
    template="""
Extract key insights from the text below.
Return the result strictly in JSON with keys:
- "summary"
- "keywords" (as a list)
- "sentiment" (positive/neutral/negative)

Text:
{text}

Return ONLY valid JSON. No explanations.
"""
)

# Sample input text
sample_text = """
Python is widely used for data science because of its rich ecosystem
of libraries such as NumPy, Pandas, and Scikit-learn.
It is also easy to learn and has a strong community.
"""

# Format prompt for the model
prompt_json = json_template.format(text=sample_text)


# ------------------------------------------------------------
# Step 2: Call OpenAI model (Chat Completions API)
# ------------------------------------------------------------
response = client.chat.completions.create(
    model="gpt-4o-mini",       # Latest lightweight OpenAI model
    messages=[{"role": "user", "content": prompt_json}]
)

raw_output = response.choices[0].message.content
print("Raw Model Output:\n", raw_output, "\n")


# ------------------------------------------------------------
# Step 3: Parse JSON safely
# ------------------------------------------------------------
try:
    # Direct JSON parsing
    parsed_output = json.loads(raw_output)
    print("Parsed JSON:\n", parsed_output)

    # Accessing individual fields
    print("\nSummary:", parsed_output["summary"])
    print("Keywords:", parsed_output["keywords"])
    print("Sentiment:", parsed_output["sentiment"])

except json.JSONDecodeError:
    print("Model returned invalid JSON. Attempting to fix...")

    # --------------------------------------------------------
    # Optional fallback: try extracting JSON from inside backticks
    # --------------------------------------------------------
    cleaned = raw_output.strip().split("```")[-1]
    try:
        parsed_output = json.loads(cleaned)
        print("Fixed JSON:\n", parsed_output)
    except:
        print("Could not parse JSON. Manual review needed.")
````

#### With parser

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()

# ------------------------------------------------------------
# Step 1: Model + Parser
# ------------------------------------------------------------
# llm    = ChatOpenAI(model="gpt-4o-mini", api_key=os.getenv("OPENAI_API_KEY"))
parser = JsonOutputParser()

# ------------------------------------------------------------
# Step 2: Prompt Template
# ------------------------------------------------------------
prompt = PromptTemplate(
    input_variables=["text"],
    template="""
Extract key insights from the text below.
Return the result strictly in JSON with keys:
- "summary"
- "keywords" (as a list)
- "sentiment" (positive/neutral/negative)

Text:
{text}

{format_instructions}
""",
    partial_variables={"format_instructions": parser.get_format_instructions()}
)

# ------------------------------------------------------------
# Step 3: Chain (replaces manual client.chat.completions.create)
# ------------------------------------------------------------
chain = prompt | llm | parser

# ------------------------------------------------------------
# Step 4: Invoke
# ------------------------------------------------------------
sample_text = """
Python is widely used for data science because of its rich ecosystem
of libraries such as NumPy, Pandas, and Scikit-learn.
It is also easy to learn and has a strong community.
"""

result = chain.invoke({"text": sample_text})

# ------------------------------------------------------------
# Step 5: Access fields directly — no json.loads() needed
# ------------------------------------------------------------
print(type(result))               # <class 'dict'>
print("Full Output:  ", result)
print("Summary:    ", result["summary"])
print("Keywords:   ", result["keywords"])
print("Sentiment:  ", result["sentiment"])
````
