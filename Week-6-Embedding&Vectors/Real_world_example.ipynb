# Clustering with Embeddings

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

documents = [
    "Payment failed during checkout",
    "Refund not received yet",
    "Cannot login to my account",
    "Password reset not working",
    "Order delivery is delayed",
    "Where is my shipment?"
]

embeddings = model.encode(documents)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(embeddings)

for i, label in enumerate(kmeans.labels_):
    print(f"Cluster {label}: {documents[i]}")
```

# Similarity detection

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Query sentence
query = "Money was deducted but payment failed"

# Candidate documents
documents = [
    "Refund has not been processed",
    "My card was charged twice",
    "I cannot login to my account",
    "Shipment tracking is not updating",
    "Payment failed during checkout"
]

# Generate embeddings
query_embedding = model.encode([query])
doc_embeddings = model.encode(documents)

# Compute cosine similarity
similarities = cosine_similarity(query_embedding, doc_embeddings)[0]

# Sort results
results = sorted(zip(documents, similarities), key=lambda x: x[1], reverse=True)

print("\n===== SIMILARITY RESULTS =====\n")
for doc, score in results:
    print(f"{score:.4f}  -->  {doc}")
```
