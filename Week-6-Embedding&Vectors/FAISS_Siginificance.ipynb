# STEP 1 — Generate Synthetic Embeddings

We simulate a realistic scenario:
- 100,000 vectors
- 384 dimensions (like MiniLM)
- 1 query vector

```python
import numpy as np
import time
from sklearn.metrics.pairwise import cosine_similarity
import faiss
```

```python
# Step 1: Generate synthetic embeddings
np.random.seed(42)

num_vectors = 100000     # database size
dimension = 384          # embedding size

# Simulated embedding database
database = np.random.random((num_vectors, dimension)).astype('float32')

# Normalize for cosine similarity
database /= np.linalg.norm(database, axis=1, keepdims=True)

# Query vector
query = np.random.random((1, dimension)).astype('float32')
query /= np.linalg.norm(query)

print("Database shape:", database.shape)
print("Query shape:", query.shape)
```

# Step-2 - Bruteforce Search

```python
# Brute Force Search
start = time.time()

similarities = cosine_similarity(query, database)[0]
top_k = 5
top_indices = np.argsort(similarities)[-top_k:][::-1]

end = time.time()

print("Brute Force Top-5 indices:", top_indices)
print("Brute Force Time:", round(end - start, 4), "seconds")
```

# Step-3 - FAISS Search 

```python
# Create index
index = faiss.IndexFlatIP(dimension)

# Add vectors
index.add(database)

# Search
start = time.time()

distances, indices = index.search(query, top_k)

end = time.time()

print("FAISS Top-5 indices:", indices[0])
print("FAISS Time:", round(end - start, 4), "seconds")
```
