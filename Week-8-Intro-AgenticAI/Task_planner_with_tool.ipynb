```python
# Import packages

from langchain.agents import initialize_agent, Tool
from langchain.chat_models import ChatOpenAI
from langchain.agents.agent_types import AgentType
import os 
from dotenv import load_dotenv
load_dotenv()
```

```python
# Task Storage - (agent’s external memory / environment)
# Creates a memory (state) → task_list
# Defines a function to store tasks

task_list = []

def add_task(task: str):
    task_list.append(task)
    return f"Task added: {task}"
```

```python
# Tool Definition (Wraps your function into something the agent can use)
# Description is critical, because LLM reads it and decides it's next steps

tools = [
    Tool(
        name="Task Manager",
        func=add_task,
        description="Add a task to the task list"
    )
]
```

```python
# LLM Setup

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0
)
```

```python
# Agent Setup (Zero-shot ReAct)
# Zero-shot --> No example 
# ReACT -> Thought → Action → Observation → Repeat

agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True  # shows reasoning trace
)
```

```python
# -----------------------------
# Run Demo
# -----------------------------
goal = "Organise a study group for our next exam"

agent.run(f"""
Split the user's goal into actionable tasks/items and use the Task Manager tool to store them:
{goal}
""")

print("\nFinal Task List:")
for i, t in enumerate(task_list, 1):
    print(f"{i}. {t}")
```
