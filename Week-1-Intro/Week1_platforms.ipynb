```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_API_BASE")  # custom OpenAI-compatible endpoint
)

def answer_query(user_query: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": user_query}
        ],
        temperature=0.0,
    )
    return response.choices[0].message.content
```

```python
Query = "What is the capital of France?"
answer_query(Query)
```

## 2. Perplexity

- Best for Research & Fact-Checking

    - https://www.perplexity.ai/account/api/keys


```python
import os
import requests
from dotenv import load_dotenv
load_dotenv()
API_KEY = os.getenv("PERPLEXITY_API_KEY")

url = "https://api.perplexity.ai/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

payload = {
    "model": "sonar-pro",
    "messages": [
        {"role": "system", "content": "You are a research assistant."},
        {"role": "user", "content": "Latest updates on OpenAI GPT models in 2025"}
    ],
    "temperature": 0.2,
    "top_p": 0.9,
    "return_citations": True
}

response = requests.post(url, headers=headers, json=payload)
response.raise_for_status()

data = response.json()

print("Answer:\n", data["choices"][0]["message"]["content"])
print("\nCitations:\n", data.get("citations", []))
```

```python
import os
from dotenv import load_dotenv
import google.generativeai as genai
# import google.genai as genai

load_dotenv()

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

model = genai.GenerativeModel("gemini-3-flash-preview")

def answer_query(user_query: str) -> str:
    response = model.generate_content(
        user_query,
        generation_config={"temperature": 0.0}
    )
    return response.text
```

```python
Query = "Explain the theory of relativity in simple terms. 100 characters "
answer_query(Query)
```

## 4. DeepSeek

- Best for Cost-Effective Engineering

    - https://platform.deepseek.com/usage

```python
from openai import OpenAI
import os
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url="https://api.deepseek.com"
)

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain attention mechanism in transformers."}
    ],
    temperature=0.2
)

print(response.choices[0].message.content)
```
